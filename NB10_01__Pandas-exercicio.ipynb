```python
def transforma_valores(coluna):
    '''
    transforma valores das colunas wage, value e release_clause de str para numérico
    entrada: uma das colunas citadas com aspas simples, p. ex: 'wage'
    saida: uma series com os valores transformados
    '''
    teste = df_fifa[coluna].replace('(^\D)(\d*\.?\d*)(M?)(K?)', '\\1|\\2|\\3\\4', regex=True, inplace=False).str.split(pat='|', expand=True)
    teste[1] = teste[1].astype('float')
    teste[2] = teste[2].map({'M':'1000000', 'K':'1000', '':'1'}).astype('float')
    return teste[1] * teste[2]
```

```python
def transforma_valores_v2(coluna): # esta versão não consegue contornar os NaN, apresenta erro.
    '''
    transforma colunas wage, value e release_clause de str para numérico
    entrada: uma das colunas citadas com aspas simples, p. ex: 'wage'
    saida: uma series com os valores transformados
    '''
    teste = df_fifa[coluna].replace({'^\D(\d*\.?\d*)':'\\1 * 1','M$':'* 1000000','K$':'* 1000'}, regex=True).apply(eval)
    return teste
```

```python
def get_title(name):
    if '.' in name:
        return name.split(',')[1].split('.')[0].strip()
    else:
        return 'Unknown'

def title_map(title):
    if title in ['Mr', 'Ms']:
        return 1
    elif title in ['Master']:
        return 2
    elif title in ['Ms','Mlle','Miss']:
        return 3
    elif title in ["Mme", "Ms", "Mrs"]:
        return 4
    elif title in ["Jonkheer", "Don", "Sir", "the Countess", "Dona", "Lady"]:
        return 5
    elif title in ["Capt", "Col", "Major", "Dr", "Rev"]:
        return 6
    else:
        return 7
```

## Relação causa --> efeito

A função a seguir nos ajudará com o Data Visualization, cruzando a variável-resposta 'survived' com qualquer outra passada à função:

```python
def Avalia_Taxa_Sobrevivencia(df, column):
    title_xt = pd.crosstab(df[column], df['survived'])
    print(pd.crosstab(df[column], df['survived'], margins=True))
    title_xt_pct = title_xt.div(title_xt.sum(1).astype(float), axis=0)
    
    title_xt_pct.plot(kind='bar', stacked=True, title='Taxa de Sobrevivência dos Passageiros', 
                      color= ['r', 'g'])
    plt.xlabel(column)
    plt.ylabel('Taxa de Sobrevivência')
    plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.05),shadow=True, ncol=2)
    plt.show()

def Catplot_Graph(x, y, hue= 'survived', col= None):
    plt.rcdefaults()
    g= sns.catplot(x= x, y= y, hue= hue, palette={'Died':'red','Survived':'blue'}, col= col, data=df, kind= 'bar', height=4, aspect=.7)
    plt.show()
```
